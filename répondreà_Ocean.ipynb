```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import product
from scipy.stats import ttest_1samp

# Load your CSV file
file_path = r'C:\Users\Subhadeep Maishal\Music\JUP\test_data\CBN_INPUT.csv'
df = pd.read_csv(file_path, parse_dates=['DATE'])

# List of oceanic indices
oceanic_indices = ['AAO', 'AO', 'DMI', 'MEIv2', 'NAO', 'ONI', 'PDO', 'SOI', 'TNA', 'TSA', 'WHWP', 'WP']

# Dictionary to store mean responses and significance
mean_responses = {}
significance = {}

# Significance level (alpha)
alpha = 0.05

# Simulate linear responses for all combinations
for (row, col) in product(range(len(oceanic_indices)), repeat=2):
    perturbed_index = oceanic_indices[row]
    response_index = oceanic_indices[col]

    if perturbed_index != response_index:
        # Introduce a small perturbation in the perturbed index
        perturbation = 0.1 * np.random.randn(len(df))
        df[perturbed_index + '_perturbed'] = df[perturbed_index] + perturbation

        # Simulate a linear response of the response index to the perturbation
        linear_coefficient = 0.5  # Adjust as needed
        df[response_index + f'_linear_response_to_{perturbed_index}'] = linear_coefficient * df[perturbed_index + '_perturbed']

        # Store the mean of the linear response
        mean_response = df[response_index + f'_linear_response_to_{perturbed_index}'].mean()
        mean_responses[(response_index, perturbed_index)] = mean_response

        # Perform t-test for significance
        t_stat, p_value = ttest_1samp(df[response_index + f'_linear_response_to_{perturbed_index}'], 0)
        significance[(response_index, perturbed_index)] = p_value < alpha

# Create a matrix of mean responses
mean_responses_matrix = np.array([[mean_responses.get((response_index, perturbed_index), 0) 
                                   for perturbed_index in oceanic_indices] 
                                  for response_index in oceanic_indices])

# Create a matrix of significance
significance_matrix = np.array([[significance.get((response_index, perturbed_index), False) 
                                  for perturbed_index in oceanic_indices] 
                                 for response_index in oceanic_indices])

# Set a custom color palette
custom_palette = sns.light_palette("navy", as_cmap=True)

# Plot the mean linear responses with a heatmap
plt.figure(figsize=(14, 8))

# Plot the mean linear responses with a heatmap
plt.subplot(1, 2, 1)
sns.heatmap(mean_responses_matrix, cmap=custom_palette, annot=True, fmt='.2f', square=True,
            xticklabels=oceanic_indices, yticklabels=oceanic_indices, cbar_kws={'label': 'Mean Linear Response'})
plt.title('Mean Linear Responses Matrix', fontsize=16)

# Plot the significance matrix with a heatmap
plt.subplot(1, 2, 2)
sns.heatmap(significance_matrix, cmap='coolwarm', annot=True, fmt='d', square=True,
            xticklabels=oceanic_indices, yticklabels=oceanic_indices, cbar=False)
plt.title('Significance Matrix', fontsize=16)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import product
from scipy.stats import ttest_1samp

# Load your CSV file
file_path = r'C:\Users\Subhadeep Maishal\Music\JUP\test_data\CBN_INPUT.csv'
df = pd.read_csv(file_path, parse_dates=['DATE'])

# List of oceanic indices
oceanic_indices = ['AAO', 'AO', 'DMI', 'MEIv2', 'NAO', 'ONI', 'PDO', 'SOI', 'TNA', 'TSA', 'WHWP', 'WP']

# Dictionary to store mean responses and significance
mean_responses = {}
significance = {}

# Significance level (alpha)
alpha = 0.5

# Significance threshold (close to zero)
threshold = 0.1

# Simulate linear responses for all combinations
for (row, col) in product(range(len(oceanic_indices)), repeat=2):
    perturbed_index = oceanic_indices[row]
    response_index = oceanic_indices[col]

    if perturbed_index != response_index:
        # Introduce a small perturbation in the perturbed index
        perturbation = 0.1 * np.random.randn(len(df))
        df[perturbed_index + '_perturbed'] = df[perturbed_index] + perturbation

        # Simulate a linear response of the response index to the perturbation
        linear_coefficient = 0.5  # Adjust as needed
        df[response_index + f'_linear_response_to_{perturbed_index}'] = linear_coefficient * df[perturbed_index + '_perturbed']

        # Store the mean of the linear response
        mean_response = df[response_index + f'_linear_response_to_{perturbed_index}'].mean()
        mean_responses[(response_index, perturbed_index)] = mean_response

        # Perform t-test for significance
        t_stat, p_value = ttest_1samp(df[response_index + f'_linear_response_to_{perturbed_index}'], 0)
        significance[(response_index, perturbed_index)] = abs(mean_response) > threshold and p_value < alpha

# Create a matrix of mean responses
mean_responses_matrix = np.array([[mean_responses.get((response_index, perturbed_index), 0) 
                                   for perturbed_index in oceanic_indices] 
                                  for response_index in oceanic_indices])

# Create a matrix of significance
significance_matrix = np.array([[significance.get((response_index, perturbed_index), False) 
                                  for perturbed_index in oceanic_indices] 
                                 for response_index in oceanic_indices])

# Set a custom color palette
custom_palette = sns.light_palette("navy", as_cmap=True)

# Plot the mean linear responses with a heatmap
plt.figure(figsize=(14, 8))

# Plot the mean linear responses with a heatmap
sns.heatmap(mean_responses_matrix, cmap=custom_palette, annot=True, fmt='.2f', square=True,
            xticklabels=oceanic_indices, yticklabels=oceanic_indices, cbar_kws={'label': 'Mean Linear Response'})

# Overlay markers for non-significant responses
for i in range(len(oceanic_indices)):
    for j in range(len(oceanic_indices)):
        response_index = oceanic_indices[i]
        perturbed_index = oceanic_indices[j]

        if perturbed_index != response_index and not significance[(response_index, perturbed_index)]:
            plt.text(j + 0.5, i + 0.5, '+', color='red', fontsize=16, ha='center', va='center')

# Customize the plot
plt.title('Mean Linear Responses Matrix with Significance when alpha = 0.5', fontsize=16)
plt.xlabel('Perturbed Index', fontsize=14)
plt.ylabel('Response Index', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```
